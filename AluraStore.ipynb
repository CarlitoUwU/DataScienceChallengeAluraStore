### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación

```python
# Análisis de facturación: ingreso total por tienda
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()

# Mostrar los resultados de forma ordenada
print("Ingreso total por tienda:")
print(f"Tienda 1: ${ingreso_tienda_1:,.2f}")
print(f"Tienda 2: ${ingreso_tienda_2:,.2f}")
print(f"Tienda 3: ${ingreso_tienda_3:,.2f}")
print(f"Tienda 4: ${ingreso_tienda_4:,.2f}")
```

## Análisis de Facturación: Ingreso Total por Tienda

El gráfico de barras muestra los ingresos totales generados por cada tienda durante el período analizado. Los valores están expresados en **pesos colombianos (COP)**.

- **Tienda 1** es la que más ingresos generó, superando los $1,150 millones.

- **Tienda 4** es la que menos ingresos obtuvo, con poco más de $1,038 millones.

- Aunque las diferencias entre las tiendas no son extremas, hay una **brecha de más de $100 millones COP** entre la tienda con mayores ingresos (Tienda 1) y la de menores ingresos (Tienda 4).


```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda_1, ingreso_tienda_2, ingreso_tienda_3, ingreso_tienda_4]

# Crear la gráfica de barras
plt.figure(figsize=(8, 5))
plt.bar(tiendas, ingresos, color='seagreen')
plt.title('Ingreso Total por Tienda')
plt.ylabel('Ingresos (COP)')
plt.xlabel('Tiendas')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()

# Mostrar etiquetas de valor sobre cada barra
for i, valor in enumerate(ingresos):
  plt.text(i, valor + 5000000, f"${valor:,.0f}", ha='center', va='bottom')

plt.show()
```

# 7. Análisis del desempeño geográfico
En esta sección se analiza la distribución geográfica de las ventas de las cuatro tiendas utilizando los datos de latitud y longitud proporcionados. El objetivo es identificar patrones espaciales que puedan ayudar a comprender mejor el rendimiento comercial según la ubicación.

---

### Visualización Individual de las Tiendas

A continuación, se presentan los mapas de calor generados para cada tienda por separado. Se utilizaron las coordenadas geográficas disponibles en el conjunto de datos para representar la concentración de ventas.

> ⚠️ **Nota importante**: Al revisar los datos, se encontró que las cuatro tiendas comparten un conjunto limitado de coordenadas geográficas. Esto significa que los registros no reflejan ubicaciones reales dispersas, sino ubicaciones fijas asociadas a cada tienda. Por lo tanto, la intensidad de los puntos en los mapas refleja la cantidad de ventas y no una distribución espacial auténtica.


```python
def agrupar_coordenadas(df):
  return df.groupby(['lat', 'lon']).size().reset_index(name='count')

# Agrupar cada tienda por coordenadas
t1 = agrupar_coordenadas(tienda)
t2 = agrupar_coordenadas(tienda2)
t3 = agrupar_coordenadas(tienda3)
t4 = agrupar_coordenadas(tienda4)

# Crear figura y ejes
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Lista para acceder a datos fácilmente
datos = [(t1, 'Tienda 1'), (t2, 'Tienda 2'), (t3, 'Tienda 3'), (t4, 'Tienda 4')]

# Graficar
for ax, (data, title) in zip(axs.flat, datos):
  scatter = ax.scatter(data['lat'], data['lon'],
                        c=data['count'], cmap='viridis',
                        s=40, alpha=0.8)
  ax.set_title(title)
  ax.set_xlabel('Latitud')
  ax.set_ylabel('Longitud')
  ax.grid(True)
  # Añadir colorbar a cada gráfico
  cbar = plt.colorbar(scatter, ax=ax)
  cbar.set_label('Cantidad de ventas')

# Ajustar diseño
plt.tight_layout()
plt.show()
```

### Visualización Interactiva con Folium

Para complementar el análisis, se generó un mapa interactivo con la biblioteca **Folium**, mostrando los puntos de calor de todas las tiendas simultáneamente.

Esta visualización permite observar la concentración de ventas agrupadas por ubicación. Sin embargo, al igual que con los mapas individuales, las coordenadas no muestran una dispersión real.

```python
import folium
from folium.plugins import HeatMap

# Crear un mapa centrado en un punto intermedio
mapa = folium.Map(location=[4.60971, -74.08175], zoom_start=6)

# Agregar puntos de calor para cada tienda
heat_data_1 = [[row['lat'], row['lon']] for index, row in tienda.iterrows()]
heat_data_2 = [[row['lat'], row['lon']] for index, row in tienda2.iterrows()]
heat_data_3 = [[row['lat'], row['lon']] for index, row in tienda3.iterrows()]
heat_data_4 = [[row['lat'], row['lon']] for index, row in tienda4.iterrows()]

# Agregar capas de Heatmap
HeatMap(heat_data_1, radius=10, blur=15).add_to(mapa)
HeatMap(heat_data_2, radius=10, blur=15).add_to(mapa)
HeatMap(heat_data_3, radius=10, blur=15).add_to(mapa)
HeatMap(heat_data_4, radius=10, blur=15).add_to(mapa)

# Mostrar el mapa
mapa
```
